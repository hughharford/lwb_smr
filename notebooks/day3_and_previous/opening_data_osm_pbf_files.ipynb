```python
class HotelHandler(osmium.SimpleHandler):
    '''
        Demonstrates how to read basics from an OSM .pbf file. 
        In this example, loads from the .pbf all the hotel names, and prints as a sorted list
        
        see ref: https://docs.osmcode.org/pyosmium/latest/intro.html
    '''
    def __init__(self):
        super(HotelHandler, self).__init__()
        self.hotels = []

    def node(self, o):
        if o.tags.get('tourism') == 'hotel' and 'name' in o.tags:
            self.hotels.append(o.tags['name'])


h = HotelHandler()
h.apply_file(osmfile, locations=True, idx='flex_mem')

# PRINT HERE will show all the hotel names in the PBF you loaded
# print(sorted(h.hotels))
```
